```python
import requests
```

# requests 로 웹페이지 접속

```python
url = "https://www.naver.com/"

requests.get(url)
```

```python
url = "https://www.naver.com/a"

res = requests.get(url)
```

```python
print(res)
```

```python
# 응답코드 가져오기
print(res.status_code)
```

```python
# 헤더 가져오기
url = "https://www.naver.com/"

res = requests.get(url)

print(res)
print(res.headers)
```

```python
# html코드 보기
print(res.text)
```

```python
# 바이너리 형태로 값을 가져오기 때문에 res.text 에서 한글이 깨지는 현상을 방지할 수 있음
print(res.content)
```

# urllib 으로 웹페이지 접속

```python
from urllib.request import urlopen
```

```python
page = urlopen(url)
```

```python
print(page)
```

```python
print(page.code)
print(page.headers)
print(page.url)
```

```python
# requests의 content와 동일한 기능
print(page.read())
```
