# 데이터 시각화

- 데이터를 분석한 결과 쉽게 이해할 수 있도록 표현하여 전달하는 것

- 데이터 시각화가 필요한 이유
    - 많은 양의 데이터를 한 눈에 살펴볼 수 있음
    - 전문 지식이 없더라도 누구나 쉽게 데이터를 인지하고 사용 가능
    - 단순한 데이터의 요약이나 나열보다 더 정확한 데이터 분석 결과 얻을 수 있음
    - 단순한 데이터에서는 알 수 없었던 중요한 정보 파악 가능
      
- 파이썬의 시각화 라이브러리
    - matplotlib ('https://matplotlib.org/')
        - 파이썬에서 가장 많이 사용하는 시각화 라이브러리
        - 판다스의 데이터프레임을 바로 시각화할 때 사용
        - 데이터 분석 이전의 데이터 이해를 위한 시각화 또는 데이터 분석 이후의 결과를 시작하기 위해 사용
    
    - Seaborn ('https://seaborn.pydata.org/')
        - matplotlib을 기반으로 색 테마, 차트 기능 들을 추가해주는 라이브러리
        - matplotlib과 함께 사용하며 히트맵, 카운트 플롯 등을 제공
          
    - floium
        - 지도 데이터를 이용하여 위치 정보를 시각화하는 라이브러리
        - 자바스크립트 기반의 인터랙티브 그래프 그릴 수 있음
    
    - plotly
        -  인터랙티브 그래프를 그려주는 라이브러리

# 막대 그래프
- 집단별 차이를 표현할 때 주로 사용
    - 크고 작음의 차이를 한 눈에 파악하기 위해
    - 예) 수량의 많고 적음 비교, 변화된 양에 대한 일별, 월별, 연별 통계 등의 비교

- 가독성 면에서 막대그래프는 일반적으로 항목의 개수가 적으면 가로 막대가 보기 좋고, 항목이 많으면 세로 막대가 보기 편함

```python
x = np.arange(3)
years = ["2017", "2018", "2019"]
values = [100, 400, 900]

plt.bar(x, values)
plt.xticks(x, years)
plt.show()
```

```python
x = np.arange(3)
years = ["2017", "2018", "2019"]
values = [100, 400, 900]

plt.bar(x, values, width=0.6)
plt.show()
```

```python
x = np.arange(3)
years = ["2017", "2018", "2019"]
values = [100, 400, 900]

plt.bar(x, values, width=0.6, align = "edge", color = "springgreen",
        edgecolor="gray", linewidth=3)
plt.show()
```

```python
x = np.arange(3)
years = ["2017", "2018", "2019"]
values = [100, 400, 900]

plt.bar(x, values, width=0.6, align = "edge", color = "springgreen",
        edgecolor="gray", linewidth=3, tick_label = years)
plt.show()
```

# 산점도

- 서로 다른 두 연속형 변수 사이의 관계를 나타내며 연속형 변수의 상관성을 확인할 때 산점도 그래프 사용
    - 예) 나이와 소득에 대한 상호 관련성 파악
      
- 산점도에 표시되는 각 점들은 자료의 관측값을 의미하고, 각 점의 위치는 관측값이 가지는 x축, y축 변수의 값으로 결정

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)

plt.scatter(x, y)
plt.show()
```

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
color = np.random.rand(N)
area = (30 * np.random.rand(N)) ** 2

plt.scatter(x, y, s = area, c = color, alpha = 0.5)
plt.show()
```

# 산점도 그래프

```python
ax = plt.subplots()
ax = sns.regplot(x = "total_bill", y = "tip", data = tips)

ax.set_xlabel("Total Bill")           
ax.set_ylabel("Tip")
plt.show()
```

```python
ax = plt.subplots()
ax = sns.regplot(x = "total_bill", y = "tip", data = tips, fit_reg = False)
ax.set_xlabel("Total Bill")           
ax.set_ylabel("Tip")
plt.show()
```

# 산점도와 히스토그램 한 번에 그리기

```python
joint = sns.jointplot( x = "total_bill", y = "tip", data = tips)

plt.show()
```

- 산점도는 점이 겹칠 경우 구분하기 어려움 -> 육각 그래프 사용

- 육각 그래프 : 2차원 표면에 육각형으로 데이터를 쌓아 표현, 특정 데이터의 개수가 많아지면 진한 색으로 표현

```python
joint = sns.jointplot(x = "total_bill", y = "tip", data = tips, kind = "hex")

plt.show()
```

## 이차원 밀집도

```python
ax = plt.subplots()
ax = sns.kdeplot(data = tips, x = "total_bill", y = "tip", fill = True)

ax.set_xlabel("Total_Bill")
ax.set_ylabel("Tip")

plt.show()
```
