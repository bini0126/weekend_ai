## Genre 결측치처리

```python
df_2020_2['genre'] = df_2020_2['genre'].fillna('Unknown')
df_2020_2['genre_list'] = df_2020_2['genre'].apply(lambda x: [g.strip() for g in x.split(',')] if x else [])
```

```python
df_2020_2.head()
```

```python
from itertools import chain

# 모든 영화의 genre_list를 합쳐서 유일 장르 set 생성
all_genres = set(chain.from_iterable(df_2020_2['genre_list']))
print(all_genres)
```

```python
# 각 유일 장르에 대해 원핫 컬럼 생성
for genre in all_genres:
    # genre가 해당 영화의 genre_list에 있으면 1, 아니면 0
    df_2020_2[f'genre_{genre}'] = df_2020_2['genre_list'].apply(lambda x: int(genre in x))
```

```python
df_2020_2.head()
```

## synopsis 결측치 처리

```python
#df_2020_2.drop(columns=['synopsis'], inplace=True)
```

```python
if 'synopsis' in df_2020_2.columns:
    df_2020_2.drop(columns=['synopsis'], inplace=True)
```

```python
# 1️⃣ 필요 라이브러리 불러오기
from sklearn.feature_extraction.text import TfidfVectorizer

# 2️⃣ tags 컬럼 결측치 확인 및 제거
df_2020_2_tags = df_2020_2.dropna(subset=['tags']).copy()

# 3️⃣ tags 문자열 전처리: 콤마(,) → 공백으로 변환하여 TF-IDF가 단어 구분 가능하도록 함
df_2020_2_tags['tags_text'] = df_2020_2_tags['tags'].apply(lambda x: x.replace(',', ' '))

# 4️⃣ TF-IDF 벡터라이저 생성
# (상위 50개의 가장 많이 등장하는 단어만 feature로 추출)
vectorizer = TfidfVectorizer(max_features=50)

# 5️⃣ TF-IDF 벡터화 적용
tfidf_matrix = vectorizer.fit_transform(df_2020_2_tags['tags_text'])

# 6️⃣ 벡터 결과를 데이터프레임으로 변환 (컬럼명: 추출된 단어 리스트)
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=vectorizer.get_feature_names_out())

# 7️⃣ 원본 데이터와 TF-IDF 결과 병합
df_2020_2_with_tfidf = pd.concat([df_2020_2_tags.reset_index(drop=True), tfidf_df], axis=1)

# 8️⃣ 결과 일부 확인
print(df_2020_2_with_tfidf.head())
```
