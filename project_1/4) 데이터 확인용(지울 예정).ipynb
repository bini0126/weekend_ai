```python
import pandas as pd
from pytrends.request import TrendReq
```

```python
df_title = df['title'].values.tolist()
df_title
```

```python
# 1. Pytrends 설정
pytrends = TrendReq(hl='en', tz=540)

# 2. 키워드 목록 (최대 5개)
kw_list = ['Sonic the Hedgehog', 'Pizza Dare 1', 'Bosomy Mom', "Stepmom's Desire"]

# 3. 검색 관심도 요청 (예: 2020~2022년)
pytrends.build_payload(
    kw_list, 
    cat=0, 
    timeframe='2020-01-01 2025-5-31', 
    geo='KR', 
    gprop=''
)

# 4. 주간 데이터 가져오기
data = pytrends.interest_over_time()

# 5. isPartial 제거
data = data.drop(columns=['isPartial'])

# 6. 월별 평균으로 리샘플링
monthly_data = data.resample('M').mean()

# 7. 결과 출력
print(monthly_data)
```

```python
from pytrends.request import TrendReq
import pandas as pd
import time

# Pytrends 세팅
pytrends = TrendReq(hl='en', tz=540)
kw_list = df_title

# 2. 결과를 저장할 리스트
all_results = []

# 3. 5개씩 나눠서 반복 요청
for i in range(0, len(kw_list), 5):
    batch = kw_list[i:i+5]
    
    try:
        # build_payload 호출
        pytrends.build_payload(batch, cat=0, timeframe='2020-01-01 2022-12-31', geo='KR', gprop='')

        # 데이터 가져오기
        data = pytrends.interest_over_time().drop(columns=['isPartial'])

        # 월별 평균 리샘플링
        monthly = data.resample('M').mean()

        # 키워드 이름으로 열 이름 붙이기
        all_results.append(monthly)
        
        time.sleep(1)  # 구글 요청 제한 피하기 위해 잠시 대기

    except Exception as e:
        print(f"Error on batch {i//5 + 1}: {batch}")
        print(e)

# 4. 전체 결과 합치기 (컬럼 기준 병합)
final_result = pd.concat(all_results, axis=1)

# 5. 결과 확인
print(final_result)
```
